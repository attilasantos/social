```python
# Substitui '\n' por espaço e remove espaços duplicados
nasc.columns = nasc.columns.str.replace('\n', ' ', regex=False).str.strip().str.replace(r'\s+', ' ', regex=True)
```

```python
nasc
```

```python
# Substitui "-" por 0
nasc.replace("-", 0, inplace=True)

# Converte colunas numéricas (se necessário)
for col in nasc.columns:
    nasc[col] = pd.to_numeric(nasc[col], errors='ignore')
```

```python
nasc
```

```python
# Lista de regiões + extras que você quer manter
regioes_e_outros = [
    'Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro-Oeste',
    'Estrangeiro', 'Brasil, sem especificação de Unidade da Federação'
]

# Lista oficial dos 27 estados brasileiros
estados = [
    'Acre', 'Alagoas', 'Amapá', 'Amazonas', 'Bahia', 'Ceará', 'Distrito Federal', 'Espírito Santo',
    'Goiás', 'Maranhão', 'Mato Grosso', 'Mato Grosso do Sul', 'Minas Gerais', 'Pará', 'Paraíba',
    'Paraná', 'Pernambuco', 'Piauí', 'Rio de Janeiro', 'Rio Grande do Norte', 'Rio Grande do Sul',
    'Rondônia', 'Roraima', 'Santa Catarina', 'São Paulo', 'Sergipe', 'Tocantins'
]

# Junta tudo
linhas_desejadas = estados + regioes_e_outros

# Filtra linhas que estão na lista
nasc_filtrado = nasc[nasc['Lugar de residência da mãe'].isin(linhas_desejadas)].copy()

# Remove cidades com nome igual ao estado (São Paulo, Rio de Janeiro) mantendo o maior valor (o estado)
nasc_filtrado = nasc_filtrado.sort_values('Total de Registros', ascending=False)
nasc_filtrado = nasc_filtrado.drop_duplicates(subset=['Lugar de residência da mãe'], keep='first').reset_index(drop=True)
```

```python
nasc_filtrado
```

```python
estado_para_regiao = {
    'Acre': 'Norte', 'Amapá': 'Norte', 'Amazonas': 'Norte', 'Pará': 'Norte',
    'Rondônia': 'Norte', 'Roraima': 'Norte', 'Tocantins': 'Norte',

    'Alagoas': 'Nordeste', 'Bahia': 'Nordeste', 'Ceará': 'Nordeste',
    'Maranhão': 'Nordeste', 'Paraíba': 'Nordeste', 'Pernambuco': 'Nordeste',
    'Piauí': 'Nordeste', 'Rio Grande do Norte': 'Nordeste', 'Sergipe': 'Nordeste',

    'Espírito Santo': 'Sudeste', 'Minas Gerais': 'Sudeste',
    'Rio de Janeiro': 'Sudeste', 'São Paulo': 'Sudeste',

    'Paraná': 'Sul', 'Santa Catarina': 'Sul', 'Rio Grande do Sul': 'Sul',

    'Distrito Federal': 'Centro-Oeste', 'Goiás': 'Centro-Oeste',
    'Mato Grosso': 'Centro-Oeste', 'Mato Grosso do Sul': 'Centro-Oeste',
}
```

```python
nasc_estados = nasc[nasc['Lugar de residência da mãe'].isin(estados)].copy()
```

```python
nasc_estados = nasc[nasc['Lugar de residência da mãe'].isin(estados)].copy()

# Remove duplicatas com nome igual ao estado (cidade com mesmo nome)
nasc_estados = nasc_estados.sort_values('Total de Registros', ascending=False)
nasc_estados = nasc_estados.drop_duplicates(subset='Lugar de residência da mãe', keep='first').reset_index(drop=True)
```

```python
nasc_estados['Região'] = nasc_estados['Lugar de residência da mãe'].map(estado_para_regiao)

# Total de registros por região
totais_por_regiao = nasc_estados.groupby('Região')['Total de Registros'].transform('sum')

# Porcentagem
nasc_estados['% de nascimentos na Região'] = (
    nasc_estados['Total de Registros'] / totais_por_regiao * 100
).round(2)
```

```python
nasc_estados
```

```python
nasc_estados['% de nascimentos no Brasil'] = (
    nasc_estados['Total de Registros'] / nasc_estados['Total de Registros'].sum()
).round(4) * 100
```
